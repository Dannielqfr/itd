```python
# Leer el archivo Excel
dataframe = pd.read_excel("./1_report.xls", sheet_name="Reporte de Asistencia")

# Renombrar columnas
dataframe.rename(columns={
    'Reporte de Eventos de Asistencia': 'Día 1',
    **{f"Unnamed: {i}": f"Día {i+1}" for i in range(1, 31)}
}, inplace=True)

# Rellenar valores NaN con cadenas vacías
dataframe.fillna("", inplace=True)

# Procesar datos
resultados = []
errores = []

for index, xls in dataframe.iterrows():
    if index <= 2:  # Saltar encabezados
        continue
    if xls[0] == "ID:":
        nombre = xls[10]
        print("Procesando:", nombre)
        filas = dataframe.iloc[index + 1]
        for inde, col in filas.iteritems():
            if pd.isnull(col) or col == "":
                resultados.append({"Nombre": nombre, "Día": inde, "Estado": "FERIADO", "Hora": None})
                continue
            if (len(col) % 5) != 0:
                errores.append({"Nombre": nombre, "Día": inde, "Valor": col})
                continue
            array = [col[i:i+5] for i in range(0, len(col), 5)]
            for substring in array:
                resultados.append({"Nombre": nombre, "Día": inde, "Estado": "ASISTENCIA", "Hora": substring})

# Crear DataFrame de resultados
df_resultados = pd.DataFrame(resultados)
df_errores = pd.DataFrame(errores)

# Exportar resultados a Excel
with pd.ExcelWriter("resultados_asistencia.xlsx") as writer:
    df_resultados.to_excel(writer, sheet_name="Resultados", index=False)
    df_errores.to_excel(writer, sheet_name="Errores", index=False)
```

```python
def formato12h(a):
    if len(a)==0:
        return "No Registró"
    if "rro" in a:
        return a
    if len(a)!=5:
        #ELVIS aquí validar la diferencia de minutos en caso la hora sea similar
        return f'Error: {a}'
    hora = a[:a.find(":")]
    minuto = a[a.find(":"):]
    final=""
    #print(hora)
    if int(hora)<13:
        final=f'{hora}{minuto} a.m.'
    else:
        mihora =f'{int(hora)-12}'
        if int(hora) <22:
            mihora=f'0{int(hora)-12}'
        final=f'{mihora}{minuto} p.m.'
    return final
```

```python
def stringToArrayHoras(a):
    if (len(a)%5)!=0:
        return "No calculable"
    arrays = []
    counter = 0
    hora=""
    for i,char in enumerate(a):
        hora+=char
        counter+=1
        if (counter%5)==0:
            arrays.append(hora)
            counter=0
            hora=""
    return arrays
```

```python
def evaluarSiSonHorasCorrectas(arr):
    isCorrectHour = True
    for hour in arr:
        hora=int(hour[:2])
        if hora>=11 and hora<=16:
            isCorrectHour = False
            break
    return isCorrectHour
```

```python
def desvEstandar(arr):
    if arr=="No calculable":
        return ""
    if len(arr)==0:
        return ""
    arr_minutos=[]
    primer=""
    ultimo=""
    counter = 0
    inicio=0
    fin=0
    for hour in arr:
        ultimo=hour
        hora=int(hour[:2])
        minuto=int(hour[3:])
        total_minutos=hora*60+minuto
        arr_minutos.append(total_minutos)
        if counter==0:
            primer=hour
            inicio=total_minutos
        counter+=1
        ultimo=hour
        fin=total_minutos
    desv=numpy.sqrt(numpy.var(arr_minutos,ddof=1))
    if evaluarSiSonHorasCorrectas(arr)==False:
        if desv>4:
            return f'Error: Refrigerio - {arr}'
        if inicio<=800:
            return primer
        return ultimo
    if desv>10:
        return f'Error: Ingreso o Salida - {arr}'
    if int(primer[:2])<=10:
        return ultimo
    return primer

#print(formato12h(desvEstandar(stringToArrayHoras("18:3718:3718:37"))))
```

```python
def arrMInutosFromArr(arr):
    if len(arr)==0:
        return "Error Minutos Diferencia"
    arr_minutos=[]
    for hour in arr:
        hora=int(hour[:2])
        minuto=int(hour[3:])
        total_minutos=hora*60+minuto
        arr_minutos.append(total_minutos)
    return arr_minutos
```

```python
def calculoRefrigerio(string):
    arr = stringToArrayHoras(string)
    if len(arr)!=2:
        return "Revisar", f'Revisar: {string}'
    arr_minutos = arrMInutosFromArr(arr)
    desv = numpy.sqrt(numpy.var(arr_minutos,ddof=1))
    if desv>=52:
        return "Revisar", f'Revisar: {string}'
    return arr[0], arr[1] 
```
